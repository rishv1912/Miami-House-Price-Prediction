```python
def remove_outliers(df,columns):
    df_filtered = df.copy()
    for col in columns:
        percentile25 = df[col].quantile(0.25)
        percentile75 = df[col].quantile(0.75)
        
        iqr = percentile75 - percentile25
        
        upper_limit = percentile75 + 1.5 * iqr
        lower_limit = percentile25 - 1.5 * iqr
        
        # df_without_outlier = df.loc[(df[col] > upper_limit) | (df[col] < lower_limit), column]
        df_filtered = df_filtered[(df_filtered[col] >= lower_limit) & (df_filtered[col] <= upper_limit)]
    return df_filtered
```

```python
new_df_without_outlier = remove_outliers(df,['LND_SQFOOT','TOT_LVG_AREA','SPEC_FEAT_VAL','RAIL_DIST','WATER_DIST','CNTR_DIST','SUBCNTR_DI','HWY_DIST','SALE_PRC'])
```

## Independence 

```python
import statsmodels.api as sm

# Fit a linear model
X = df[['LATITUDE', 'LONGITUDE', 'LND_SQFOOT', 'TOT_LVG_AREA', 'SPEC_FEAT_VAL', 
        'RAIL_DIST', 'OCEAN_DIST', 'WATER_DIST', 'CNTR_DIST', 'SUBCNTR_DI', 
        'HWY_DIST', 'age', 'avno60plus', 'month_sold', 'structure_quality']]
X = sm.add_constant(X)
model = sm.OLS(df['SALE_PRC'], X).fit()

# Durbin-Watson test statistic
dw_statistic = sm.stats.durbin_watson(model.resid)
print(f'Durbin-Watson Statistic: {dw_statistic}')
```
