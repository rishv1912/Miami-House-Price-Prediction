```python
def remove_outliers(df,column):
    percentile25 = df.quantile(0.25)
    percentile75 = df.quantile(0.75)
    
    iqr = percentile75 - percentile25
    
    upper_limit = percentile25 + 1.5 * iqr
    lower_limit = percentile75 - 1.5 * iqr
    
    df_without_outlier = df.loc[(df[column] < lower_limit) | (df[column] > upper_limit), column]
```
